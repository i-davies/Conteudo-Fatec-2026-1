# Revisão: Escalando o Perceptron com NumPy
Adicionando complexidade ao modelo do banco utilizando novos parâmetros e processamento em lote (Batch).

```python
import numpy as np

# Pesos (Weights) agora são um Vetor NumPy!
# Matriz Lógica: [Renda, Histórico, Idade, Garantia]
pesos = np.array([0.3, 0.8, 0.1, 0.5])

# Bias (Continua sendo apenas um número)
bias = -9.0  # Banco ficou ainda mais restrito neste novo modelo!
```

---

```python
# Dados de UM cliente isolado (Vetor X)
# [Renda, Histórico, Idade, Garantia]
cliente_teste = np.array([7.0, 8.0, 4.0, 6.0])

# Cálculo Z utilizando np.dot (Produto Escalar)
# Em vez de somas e multiplicações imensas soltas, delegamos o bloco matemático pra o C através do Numpy
z = np.dot(cliente_teste, pesos) + bias

# Função de Ativação (Degrau)
decisao = 1 if z >= 0 else 0
print(f"O valor Z foi {z:.2f} | Decisão: {'Aprovado' if decisao == 1 else 'Reprovado'}")
```

---

```python
# Tabela de Clientes (Matriz) - Cada linha é um cliente deferente
# Colunas: Renda, Histórico, Idade, Garantia
clientes_lote = np.array([
    [7.0, 8.0, 4.0, 6.0],  # Cliente 1 (Bom perfil)
    [3.0, 2.0, 5.0, 0.0],  # Cliente 2 (Risco altíssimo)
    [5.0, 5.0, 5.0, 10.0]  # Cliente 3 (Renda comum, mas com ótima garantia de bens)
])

# O NumPy calcula TODOS os clientes de uma vez de forma ultra-performática!
Z_lote = np.dot(clientes_lote, pesos) + bias

# Função de ativação para a Matriz inteira vetorizada
decisoes_lote = (Z_lote >= 0).astype(int)

for i, decisao in enumerate(decisoes_lote):
    status = "APROVADO" if decisao == 1 else "REPROVADO"
    print(f"Cliente {i+1} -> Z={Z_lote[i]:.2f} | Veredito: {status}")
```
