# Revisao: MinMaxScaler e OneHotEncoder
Exemplo simples para transformar dados numericos e categoricos.

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
df = pd.DataFrame([
    {"tempo_entrega": 25, "valor_pedido": 70, "tipo_entrega": "Moto"},
    {"tempo_entrega": 40, "valor_pedido": 40, "tipo_entrega": "Carro"},
    {"tempo_entrega": 30, "valor_pedido": 85, "tipo_entrega": "Bike"}
])

df
```

```python
num_cols = ["tempo_entrega", "valor_pedido"]
cat_cols = ["tipo_entrega"]

# MinMaxScaler aplica: (x - min) / (max - min)
# Por isso, o menor valor da coluna vira 0 e o maior vira 1.
preprocessor = ColumnTransformer(
    transformers=[
        ("num", MinMaxScaler(), num_cols),
        ("cat", OneHotEncoder(handle_unknown="ignore", sparse_output=False), cat_cols),
    ]
)

X = preprocessor.fit_transform(df)
X
```

```python
cat_encoder = preprocessor.named_transformers_["cat"]
cat_names = cat_encoder.get_feature_names_out(cat_cols).tolist()
feature_names = num_cols + cat_names

df_features = pd.DataFrame(X, columns=feature_names)

print("Minimos usados no scaler:")
print(df[num_cols].min())
print("\nMaximos usados no scaler:")
print(df[num_cols].max())
print("\nResultado transformado:")
df_features
```
